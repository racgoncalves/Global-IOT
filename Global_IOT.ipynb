***KNN***

```
import numpy as np
from collections import Counter
def knn(data,prd):
    dist=[]
    for grp in data:
        for pt in data[grp]:
            dist.append((np.linalg.norm(np.array(pt)-np.array(prd)),grp))
            votes= sorted(dist)[:3]
            print(votes)
            votes=[i[1] for i in votes]
    
    return Counter(votes).most_common(1)[0][0]
      
```

***Rodando o teste***

```
c=0
t=0
for gr in test_set:
    for p in test_set[gr]:
        if gr==knn(train_set,p):
            c+=1
        t+=1   
```
